# AB test umístění reklamy na webové stránce

Předpokládejme, že máme webovou stránku a chceme testovat, zda nové umístění reklamy zvýší počet kliknutí.

**Hypotéza**
- Nulová hypotéza (H0): Změna umístění reklamy nemá vliv na počet kliknutí.
- Alternativní hypotéza (H1): Změna umístění reklamy zvýší počet kliknutí.


Mějme **dvě skupiny** dat s počtem kliknutí, kde každá skupina obsahuje data z 30 dní:
- Skupina A (kontrolní skupina): Používá tradiční umístění reklamy.
- Skupina B (experimentální skupina): Reklama byla přesunuta na více prominentní pozici na stránce.

Zde jsou simulovaná data pro obě skupiny:

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Náhodně generovaná data kliknutí
np.random.seed(42)
clicks_A = np.random.poisson(lam=20, size=30)  # Kontrolní skupina
clicks_B = np.random.poisson(lam=23, size=30)  # Experimentální skupina
```

```python

# Provedení nepárového t-testu
t_stat, p_value = stats.ttest_ind(clicks_A, clicks_B)

# Vytvoření grafu pro vizualizaci dat
plt.figure(figsize=(12, 7))
plt.hist(clicks_A, alpha=0.7, label='Skupina A: Kontrolní', color='blue')
plt.hist(clicks_B, alpha=0.7, label='Skupina B: Experimentální', color='green')
plt.title('Distribuce počtu kliknutí na reklamu')
plt.xlabel('Počet kliknutí')
plt.ylabel('Frekvence')
plt.legend()
plt.grid(True)
plt.show()

# Výpis výsledků t-testu
print("t-statistika:", t_stat)
print("p-hodnota:", p_value)
```

Výsledky t-testu, které jste poskytli, ukazují:

- **t-statistika:** -3.307
- **p-hodnota:** 0.00162

### Interpretace výsledků

**P-hodnota** je klíčovým indikátorem při rozhodování o zamítnutí nulové hypotézy v kontextu A/B testování. V tomto případě:

- **P-hodnota (0.00162)** je mnohem menší než standardní úroveň významnosti 0.05. To znamená, že rozdíl mezi skupinami A a B v počtu kliknutí je statisticky významný a je velmi nepravděpodobné, že by byl způsoben náhodou.

### Doporučení

Na základě výsledků t-testu by **nulová hypotéza měla být zamítnuta**. Tento závěr naznačuje, že experimentální změna umístění reklamy skutečně měla pozitivní vliv na počet kliknutí ve srovnání s kontrolní skupinou. Tedy, experimentální skupina B, kde byla reklama umístěna na více prominentní pozici, generovala statisticky významně více kliknutí než kontrolní skupina A s tradičním umístěním reklamy.
